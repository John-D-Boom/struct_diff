```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x_one_comp(n, N):
    return 1/N/(1-n/N)

def x_two_comp(n, N):
    return (1-((n+1)/N))/(1-n/N)

N = 10
n = np.arange(0, N, 1)
plt.plot(n, x_one_comp(n, N), label='one compartment')
plt.plot(n, x_two_comp(n, N), label='two compartment')
```
