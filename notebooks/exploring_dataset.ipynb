```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
import sys
import pickle as pkl
sys.path.append("/burg/home/jb5005")

import matplotlib.pyplot as plt
```

```python
base_path = '/pmglocal/jb5005/struct_diff_data/data'
file_arr = os.listdir(base_path)
```

```python
# Initialize an empty list to store rows
rows = []

# Iterate over each file in file_arr
for file in file_arr:
    # Load the file using pickle
    with open(os.path.join(base_path, file), 'rb') as f:
        data = pkl.load(f)
    
    # Access the "sequence" attribute and compute its length
    seq_len = len(data["sequence"])
    
    # Extract the name (filename without .pkl)
    name = file.replace('.pkl', '')
    
    # Add the name and seq_len to the rows list
    rows.append({"name": name, "seq_len": seq_len})

# Create the DataFrame from the rows
df = pd.DataFrame(rows)
```

```python
df.to_csv('/burg/home/jb5005/struct_diff/data/seq_len.csv')
```

```python
seq_len_counts = df['seq_len'].value_counts().sort_index()

# Create a dot plot
sns.scatterplot(x=seq_len_counts.index, y=seq_len_counts.values, size = 5)

# Add labels and title
plt.xlabel("Sequence Length")
plt.ylabel("Count")
# Show the plot
plt.show()
```

```python
sns.ecdfplot(data=df, x="seq_len")

plt.xlabel("Sequence Length")
plt.ylabel("Cumulative Density")
# plt.title("Cumulative Density Plot of Sequence Lengths")
```
